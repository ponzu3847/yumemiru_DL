```python
#MNISTデータセットの読み込みとデータの加工
import numpy as np
from keras.utils import np_utils
from keras.datasets import mnist

(x_train,y_train),(x_test,y_test)=mnist.load_data()

x_train=x_train.reshape(60000,28,28,1)
x_train=x_train.astype("float32")
x_train/=255#0から255までのRGB値を0から1までに変換する
correct=10#0から9までの10種類
y_train=np_utils.to_categorical(y_train,correct)#1ofK符号化法 correct個の値のうち該当しているインデックスの値を1にする
#print(y_train)
x_test=x_test.reshape(10000,28,28,1)
x_test=x_test.astype("float32")
x_test/=255#0から255までのRGB値を0から1までに変換する
y_test=np_utils.to_categorical(y_test,correct)#1ofK符号化法 correct個の値のうち該当しているインデックスの値を1にする
```

```python
#畳み込みニューラルネットワークの構築
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D
from keras.layers import Activation,Dropout,Flatten,Dense
from keras.optimizers import Adam
import time

#畳み込み層の設定
model=Sequential()
model.add(Conv2D(filters=10,#フィルターの数
                 kernel_size=(3,3),#3*3のフィルターを使用
                 padding="same",#ゼロパディングを行う
                 input_shape=(28,28,1),#入力データのサイズ
                 activation="relu"#活性化関数はReLU
                ))
#出力層への入力を4次元配列から2次元に変換する
model.add(Flatten())
#出力層の設定
model.add(Dense(10,
                activation="softmax"
               ))
#オブジェクトのコンパイル
model.compile(loss="categorical_crossentropy",
              optimizer=Adam(),
              metrics=["accuracy"])
model.summary()
```

```python
startTime = time.time()
history = model.fit(x_train,
                    y_train,
                    epochs=20,
                    batch_size=100,
                    verbose=1,
                    validation_data=(x_test, y_test))
score=model.evaluate(x_test,y_test,verbose=0)
print("Test loss:",score[0])
print("Test accuracy:",score[1])
print("Computation time:{0:.3f} sec".format(time.time()-startTime))
```
