```python
#ステップ関数
import numpy as np
import matplotlib.pylab as plt#matplotlibのpylabをインポート

def step_function(x):
    return (x>0).astype(np.int)

x=np.arange(-5.0,5.0,0.01)
y=step_function(x)
plt.plot(x,
         y)
plt.ylim(-0.1,1.1)#y軸の範囲
plt.grid(True)
plt.yticks([-0.5,0.0,0.5,1.0,1.5])
plt.show()
```

```python
#シグモイド関数
import numpy as np
import matplotlib.pylab as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.arange(-5.0,5.0,0.01)
y=sigmoid(x)
plt.plot(x,
         y)
plt.ylim(-0.1,1.1)
plt.grid(True)
plt.yticks([-0.5,0.0,0.5,1.0,1.5])
plt.show()
```
